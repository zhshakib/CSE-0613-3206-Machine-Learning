Data Analysis using Pandas

```python
import pandas as pd

df = pd.read_csv("../datasets/titanic/titanic_full.csv")

# Basic Information Showing  
df.head(2)
# df.info()
# df.describe()
```

1. How many Been Survived??
2. Average Age? 
3. Average Fare? 

```python
df['Survived'].value_counts()
round(df['Age'].mean()) #Rounds the Average value! 
```

1. Who Survived most? Male or Female?
2. Which Class survived most? Rich, Middle or Poor? 

```python
df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
```

Did poor women survive more than rich men?

```python
df.groupby(['Sex', 'Pclass'])['Survived'].mean()
```

Lets See if we Can Visualize This Data Analyzed! 

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Sex", y="Survived", hue="Sex", data=df)
plt.title("Survived Rate by Gender")
plt.show()
```

Lets See can we Count Those values

```python
sns.countplot(x="Survived", hue="Sex", data=df)
plt.title("Survived Rate by Gender")
plt.show()
```

Lets See which Class survived most in visual! 

```python
sns.set_style('whitegrid')
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survived Rate by Class")
plt.show()
```
