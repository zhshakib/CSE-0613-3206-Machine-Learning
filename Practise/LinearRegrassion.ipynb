We have given a problem of Linear Regression. <br>
where in a market of various fish price<br>
```bash 
(Grams, Price/kg)<br>
(x,y)<br>
(250, 1000)<br>
(350, 1200)<br>
(500, 1500)<br>
(500, 1400)<br>
(700, 2200)<br>
```
**We have to predict the next Vendor's 600g's price**

Before we Go:
### Lets Gather The Formulas together



````python
import numpy as np

# Data Given
x = np.array([250,350,500,500,700])
y = np.array([1000,1200,1500,1400,2200])

# Let our Equation be: Y = mx + C;

# Calculate Mean:
x_mean = np.mean(x)     # Numpy makes it Easy Babe! 
y_mean = np.mean(y)     # Numpy makes it Easy Babe! 

print(x_mean, y_mean) # 460.0, 1460.0

# Calculate Slope m:
rise    = np.sum((x-x_mean) * (y-y_mean))
run     = np.sum((x-x_mean)**2)

m   = rise / run
print(m) # 2.5811965811965814

# Calculate Intercept C
C = round(y_mean - (m*x_mean))
print(C) # 273

# So our Equation is: 
# Y = 2.5811 * x + 273
# Now Lets Predict for 600g's Price:

predicted = m * 600 + C

print(round(predicted)) #  1822 Taka! 
````

### Okay Now Lets Draw Some Visuals with This? 

````python
import matplotlib.pyplot as plt

plt.scatter(x,y)
# annotate each point
for i in range(len(x)):
    plt.annotate(
        f'{y[i]}',
        (x[i], y[i]),
        textcoords="offset points",
        xytext=(-5,5),
        ha='left'
    )

x_line = np.linspace(min(x), max(x), 100)
y_line = m*x_line + C

plt.plot(x_line, y_line, label="Regression Line")

# Predicted Point  
plt.scatter(600, predicted, marker='x', s=140, label = "Prediction")

# Some Customization From ChatGPT
plt.annotate(
    f'Predicted: {predicted:.0f}',
    (600, predicted),
    textcoords="offset points",
    xytext=(0,-20),
    ha='left'
)
plt.axvline(x=600, linestyle='--', alpha=0.2)
plt.axhline(y=predicted, linestyle='--', alpha=0.2)

plt.legend()
plt.xlabel("Weight (grams)")
plt.ylabel("Price")
plt.show()
````

📊 Practice Problem: House Size vs Rent

A landlord collected this data:

```bash
Size (sq ft)	Monthly Rent ($)
500	800
700	950
900	1200
1200	1500
1500	1800

```
🎯 Your Task

```bash
1. Find the regression line
y=mx+b
2. Predict rent for:
👉 1000 sq ft house
3. Visualize it in Jupyter
* scatter plot
* regression line
* highlight prediction
* label all points

```

````python
data_x = np.array([500,700,900,1200,1500])
data_y = np.array([800,950,1200,1500,1800])

# Calculate mean
mean_x = np.mean(data_x)
mean_y = np.mean(data_y)

# Calculate Slope
rise = np.sum((data_x-mean_x) * (data_x- mean_y))
run = np.sum((data_x-mean_x)**2)

m = rise/run
# Calculate Intercept C
C = round(mean_y - (m * mean_x))

print(m, C)

# Now Predicion for 1000 sqft
predicion_x = 1000
Y = round(m*predicion_x + C)

print(Y)
````

### Now Lets Visualize this

````python

sns.scatterplot(x=data_x, y=data_y, label="Data")

x_line = np.linspace(500, 1500,100)
y_line = m * x_line + C 

plt.plot(x_line, y_line, label="Regression Line")
# Plot the Prediction line
plt.scatter(predicion_x, Y)
plt.axvline(x= predicion_x, linestyle="--", alpha=0.3)
plt.axhline(y= Y, linestyle="--", alpha=0.3)

plt.annotate(
    f'Predicted: {Y:.0f}',
    (predicion_x, Y),
    textcoords="offset points",
    xytext=(0,10),
    ha='right'
)
````
