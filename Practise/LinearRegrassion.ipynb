We have given a problem of Linear Regression. <br>
where in a market of various fish price<br>
(Grams, Price/kg)<br>
(x,y)<br>
(250, 1000)<br>
(350, 1200)<br>
(500, 1500)<br>
(500, 1400)<br>
(700, 2200)<br>
**We have to predict the next Vendor's 600g's price**

```python
import numpy as np

# Data Given
x = np.array([250,350,500,500,700])
y = np.array([1000,1200,1500,1400,2200])

# Let our Equation be: Y = mx + C;


# Calculate Mean:
x_mean = np.mean(x)     # Numpy makes it Easy Babe! 
y_mean = np.mean(y)     # Numpy makes it Easy Babe! 

print(x_mean, y_mean) # 460.0, 1460.0

# Calculate Slope m:
rise    = np.sum((x-x_mean) * (y-y_mean))
run     = np.sum((x-x_mean)**2)

m   = rise / run
print(m) # 2.5811965811965814

# Calculate Intercept C
C = round(y_mean - (m*x_mean))
print(C) # 273

# So our Equation is: 
#   Y = 2.5811 * x + 273

# Now Lets Predict for 600g's Price:

predicted = m * 600 + C

print(round(predicted)) #  1822 Taka! 
```

Okay Now Lets Draw Some Visuals with This? 

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)

x_line = np.linspace(min(x), max(x), 100)
y_line = m*x_line + C

plt.plot(x_line, y_line, label="Regression Line")

# Predicted Point  
plt.scatter(600, predicted, marker='x', s=140, label = "Prediction")
plt.legend()
plt.xlabel("Weight (grams)")
plt.ylabel("Price")
plt.show()
```
