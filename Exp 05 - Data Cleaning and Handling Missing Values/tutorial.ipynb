```python
df = pd.read_csv('../datasets/titanic/titanic_full.csv')
df.drop(columns=['Cabin', 'Lifeboat', 'Body', 'Name_wiki', 'Age_wiki'], inplace=True)

df.fillna({
    'Age' : df['Age'].median(),
    'Survived' : df['Survived'].mode()[0],
    'Fare' : df['Fare'].median(),
}, inplace=True)

df.to_csv('cleaned_data.csv', index=False)
```
