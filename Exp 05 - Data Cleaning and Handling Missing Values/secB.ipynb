```python
df = pd.read_csv('../datasets/titanic/titanic_full.csv')
df.drop(columns=['Cabin', 'Lifeboat', 'Body'], inplace=True)
df.fillna({
    'Age' : df['Age'].median(),
    'Survived' : df['Survived'].mode()[0],
    'Fare' : df['Fare'].median(),
    'Embarked' : df['Embarked'].mode()[0]
}, inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# df.to_csv('titanic_cleaned2.csv', index=False)

sns.heatmap(df.isnull(), cmap='plasma', cbar=False)
```
