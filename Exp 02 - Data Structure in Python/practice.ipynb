```python
scores = [85, 90, 78, 92, 88]
scores[3] = 1000
max_score = max(scores)
print(f"Scores are: {scores}")
print(f"Maximum score is: {max_score}")
print(f"Average score is: {sum(scores) / len(scores)}")
```

```python
# Create tuples from the existing variables
scores = [85, 90, 78, 92, 88]
numbers = [1, 2, 3, 4, 1, 5]
scores_tuple = tuple(scores)
numbers_tuple = tuple(numbers)

print(f"Scores as tuple: {scores_tuple}")
print(f"Numbers as tuple: {numbers_tuple}")

# Tuple unpacking
first, second, *rest = numbers_tuple
print(f"First: {first}, Second: {second}, Rest: {rest}")

# Tuple methods
print(f"Count of 1 in numbers_tuple: {numbers_tuple.count(1)}")
print(f"Index of 3 in numbers_tuple: {numbers_tuple.index(3)}")
```

```python
import numpy as np

numbers = [1, 2, 3, 4, 5]
arr = np.array(numbers)

print(arr)
mean = np.mean(arr)
standard_deviation = np.std(arr)

print("Mean:", mean)
print("Standard Deviation:", standard_deviation)

print(arr ** 2)
```
