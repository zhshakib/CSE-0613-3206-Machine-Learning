```python
# if elif else statement
score = 100

if score >= 80 and score <= 100:
    print("A")
elif score >= 70 and score < 80:
    print("B")
elif score >= 60 and score < 70:
    print("C")
elif score >= 0 and score < 60:
    print("F")
else:
    print("Invalid score")
```

```python
def even_or_odd(n):
    if n % 2 == 0:
        print("Even")
    else:
        print("Odd")

even_or_odd(10)
even_or_odd(11)
```
