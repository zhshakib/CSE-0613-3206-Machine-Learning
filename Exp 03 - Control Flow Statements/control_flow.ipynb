# Control Flow — Short Tutorial

This notebook contains short, runnable examples for comparison and logical operators, conditional statements, iterators and `range()`, `for` and `while` loops, and functions.

Run each code cell to experiment.

Comparison and logical operators are used to compare values and combine boolean expressions. Comparisons return `True` or `False`. Use parentheses to clarify complex logic.

```python
# Comparison & Logical Operators
x = 10
print('x > 5 ->', x > 5)
print("x == 5 or x > 8 ->", x == 5 or x > 8)
print('not (x < 0) ->', not (x < 0))
```

Notes: `if`/`elif`/`else` let you run different code paths based on conditions. Python uses indentation for blocks; evaluate conditions top-to-bottom.

```python
# Conditional Statements
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
else:
    grade = 'C'
print('grade ->', grade)
```

Notes: `range()` generates integer sequences. Use `for` to iterate the produced values or convert `range` to a list to inspect it.

```python
# Iterators & range()
for i in range(3):    # 0,1,2
    print('range(3):', i)

for i in range(1, 6, 2):  # 1,3,5
    print('range(1,6,2):', i)
```

Notes: `for` loops traverse any iterable (lists, strings, generators). Use `enumerate()` for index and `zip()` to loop multiple iterables together.

```python
# for Loop
fruits = ['apple', 'banana', 'cherry']
for f in fruits:
    print('fruit ->', f)

for idx, val in enumerate(fruits, start=1):
    print('index, value ->', idx, val)
```

Notes: `while` runs while a condition is true — avoid infinite loops by changing the condition inside the loop. Use `break` to exit and `continue` to skip.

```python
# while Loop
n = 3
while n > 0:
    print('n ->', n)
    n -= 1

# break / continue example
for i in range(1, 6):
    if i % 2 == 0:
        continue
    if i > 3:
        break
    print('odd and <=3 ->', i)
```

Notes: Functions group reusable code. Use parameters to pass data and `return` to send results back. Keep functions small and focused.

```python
# Functions

def greet(name):
    if not name:
        return 'Hello!'
    return f'Hello, {name}!'

print(greet('Ana'))
print(greet(''))
```
