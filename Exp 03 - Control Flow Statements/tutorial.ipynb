```python
# Comparison Operators - ==, !=, >, <, >=, <=
print(5 <= 5)
```

```python
# Logical Operators - and, or, not
a, b, c = 5, 6, 7
print(a < b or b > c)
```

```python
# Conditional Statements - if, elif, else
a, b, c = 5, 5, 7
if a > b:
    print("a is less than b")
elif a == b:
    print("a is equal to b")
else:
    print("a is greater than or equal to b")
```

```python
# Range in python

for i in range(10, 4, -1):
    print(i)
```

```python
# While loop in python
i = 1

while i < 11:
    print(i)
    i += 2
```

```python
# Functions in python
# f(x) = 2x + 3

def whatever(x):
    return 2 * x + 3

def greeting(name):
    return "Hello, " + name

def add(a, b):
    return a + b

print(add(4, 6))
```
