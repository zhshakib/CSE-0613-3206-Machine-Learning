```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Bar chart
years = [2021, 2022, 2023, 2024, 2025]
sells = [10, 12, 11, 9, 6]

plt.bar(years, sells, color="yellow", width=0.4, edgecolor="black")
plt.title("Sells over year")
plt.xlabel("Years")
plt.ylabel("Sells (100K)")
plt.show()
```

```python
# Pie
data = [123908, 150000, 170000, 200000, 250000]
data_labels = ["2021", "2022", "2023", "2024", "2025"]

plt.pie(data, labels=data_labels, autopct="%1.1f%%", explode=[0, 0, 0, 0, 0.1])
plt.title("Pie Chart")
plt.show()
```

```python
# Histogram
df = pd.read_csv('titanic_cleaned.csv')

plt.hist(df['Fare'], bins=20, color='yellow', edgecolor='black')
plt.show()
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

y_true = [0, 0, 2, 2, 3, 0, 2, 5]
y_pred = [0, 0, 2, 2, 3, 2, 2, 5]
class_names = ["cat", "cow", "sheep", "goat"]

cm = confusion_matrix(y_true, y_pred)

sns.heatmap(cm, annot=True, xticklabels=class_names, yticklabels=class_names)
plt.show()
```
