```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('titanic_cleaned.csv')
```

```python
f = lambda x: x**2
x = [1, 3, 4, 5, 6, 7, 8, 9, 10]
y = [f(i) for i in x]

plt.bar(x, y, color="orange", width=0.5)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Pie Chart Example
plt.pie([10, 20, 30], labels=["A", "B", "C"], autopct="%1.1f%%", explode=[0.1, 0, 0])
plt.title("Pie Chart Example")
plt.legend()
plt.show()
```

```python
# Histogram Example
data = [df['Fare']]
plt.hist(data, bins=10, color="orange", edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Confusion Matrix Example
from sklearn.metrics import confusion_matrix
y_true = [0, 1, 2, 3, 4, 5, 4, 5]
y_pred = [0, 0, 2, 3, 4, 5, 4, 4]
cm = confusion_matrix(y_true, y_pred)
plt.imshow(cm, cmap=plt.cm.plasma_r)
plt.title("Confusion Matrix Example")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.colorbar()
plt.show()
```
