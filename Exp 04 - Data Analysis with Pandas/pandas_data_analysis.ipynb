# Pandas — Basic Data Analysis

Short tutorial: create a DataFrame from a dict, load a CSV dataset, inspect and filter data, and compute simple summaries using `pandas`. Examples use the Titanic dataset.

```python
# Imports and options
import pandas as pd
```

```python
## Create DataFrame from dict
# Small example to see how DataFrame is constructed
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'city': ['NY', 'LA', 'SF']
}
df = pd.DataFrame(data)
df
```

```python
## Load Titanic CSV
# Update the path below if your workspace is located elsewhere.
path = r'D:/Courses/CSE 0613-3206-Machine Learning/datasets/titanic/titanic_full.csv'
titanic = pd.read_csv(path)
titanic.head()
```

```python
## Inspecting the data
# Shape, info, and descriptive statistics
print('shape:', titanic.shape)
print('\ninfo:')
titanic.info()

print('\nsummary (describe):')
titanic.describe(include='all')
```

```python
## Viewing & Selecting
# Select columns and rows
# Show a few useful columns
titanic[['PassengerId','Name','Sex','Age','Pclass']].head()
```

```python
## Filtering examples
# Filter female passengers and those older than 30
female_over30 = titanic[(titanic['Sex'] == 'female') & (titanic['Age'] > 30)]
female_over30[['Name','Age','Sex','Survived']].head()
```

```python
## Counts and Grouping
# Value counts and groupby summaries
print('Pclass counts:\n', titanic['Pclass'].value_counts())
print('\nSurvival rate by Pclass:\n', titanic.groupby('Pclass')['Survived'].mean())
```

```python
## Save filtered sample (optional)
# Save a small filtered dataset for later use
female_over30.to_csv('titanic_female_over30_sample.csv', index=False)
print('Saved titanic_female_over30_sample.csv')
```
